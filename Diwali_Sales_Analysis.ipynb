```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import plotly.express as px
```

```python
# read diwali data set
df = pd.read_csv('D:\Python Basics\Projects\Diwali Sales Data.csv',encoding='unicode_escape')
df
```

## Basic data cleaning and exploration

```python
## Check for columns
df.columns
```

```python
## check for missing values 
df.isnull().sum()
```

```python
## Remove unwanted column
df.drop(['Status','unnamed1'],axis = 1,inplace = True)
```

```python
df
```

```python
## Check statistcal info
df.describe()
```

```python
## replace missing values with 0
df['Amount_new']=df['Amount'].fillna(0)
```

```python
df['Amount_new'].isnull().sum()
```

```python
## drop Amount column
df.drop('Amount',axis = 1,inplace= True)
```

```python
## Change the value 1 to married and 0 to unmarries in the column marrital status
marital_status = df['Marital_Status'].tolist()
Marriage_status = []
for i in marital_status:
    if i == 0:
        Marriage_status.append('Unmarried')
    else:
        Marriage_status.append('Married')
# print(Marriage_status)
```

```python
df['Marital_status'] = Marriage_status
```

```python
## drop the old marrital_status col
df.drop('Marital_Status',axis = 1,inplace = True)
```

# Customer Segmentation

```python
## segreagte the customers as high , medium and low spenders based on amount they spend 
Amount = df['Amount_new'].tolist()
Customers_Segemnet = []
for i in Amount:
    if i>=3000:
        Customers_Segemnet.append('High Spenders')
    elif i<3000:
        Customers_Segemnet.append('Medium Spender')
    else:
        Customers_Segemnet.append('Low spender')


print(Customers_Segemnet)
df['Customer_Segment'] = Customers_Segemnet
df
```
